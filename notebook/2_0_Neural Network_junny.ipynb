Activation Function
---


### Reference:
https://medium.com/the-theory-of-everything/understanding-activation-functions-in-neural-networks-9491262884e0

Activation function란 다음과 같이 설명되어져 있는데요,

Activation functions are used to determine the firing of neurons in a neural network. Given a linear combination of inputs and weights from the previous layer, the activation function controls how we'll pass that information on to the next layer.

해석하자면 activation function은 다음 layer에 이 neuron에서 계산된 값을 전달할 지 말지(firing) 를 결정해주는 함수로 볼 수 있습니다.

이러한 activation function은 크게 2가지 특성을 지니고 있어야 하는데요

1) nonlinearity 비선형성
2) differentiability 미분가능성

이렇게 두가지 특성이 있어야 합니다.

1번의 특성은 최종적으로 비선형적인 특성을 나타내기 위해서는 선형 weight합에 선형 함수를 곱하여서 나중에 더하게 되면 결국 얻게 되는 network는 또 다른 하나의 동치되는 선형 layer와 같아지기 때문에, activation function은 반드시 비선형성을 가지고 있어야 합니다.

2번의 특성은 미분 가능성 인데요, neural network에서 weight들을 update할 때 아래에서 보여드릴 back propagation과정을 통해 update를 하며, 이 때 미분값을 이용하기 때문입니다.

### Vanishing Gradient

<img src="https://t1.daumcdn.net/cfile/tistory/2168AC3758D5CE0C33" alt="Drawing" style="width: 500px;"/>




sigmoid 함수의 경우 미분값이 위처럼 형성되며, 결국 미분 값에 의해서 update를 계속 할 수록 sigmoid 함수의 양극으로 가는 방향으로 weight들이 수렴하게 됩니다. 그럴 경우 sigmoid 함수의 미분값에서 볼 수 있듯이 양극단으로 갈 수록 0으로 미분 값이 향하기 때문에 점점 back propogation을 거듭할 수록 점점 더 weight들을 변화시키지 않게 됩니다. 




이 문제를 학자들은 1986년도부터 2006년도까지 무려 20년 동안 해결하지 못하였기에 neural network를 깊게 쌓아도 소용이 없었습니다.

이를 해결한 것이 바로 RELU 함수이며, max(0,x)라는 굉장히 단순한 형태이지만, 미분 값이 사라지지 않고 1을 계속 유지하는 특성 때문에 이 vanishing gradient 문제를 해결할 수 있었습니다.

Generalization
---

<img src="https://image.slidesharecdn.com/random-170910154045/95/-53-638.jpg?cb=1505089848" alt="Drawing" style="width: 600px;"/>

이러한 문제를 해결하기 위해서는 세가지 해결 방법이 있을수 있는데,


첫 번째 : training data를 많이 모으는 것입니다. 데이터가 많으면 training set, validation set, test set으로 나누어서 진행할 수 있고, 영역별 데이터의 크기도 커지기 때문에 overfitting 확률이 낮아집니다. 달리 생각하면 overfitting을 판단하기가 쉬워지기 때문에 정확한 모델을 만들 수 있다고 볼 수도 있습니다.

두 번째 : feature의 갯수를 줄이는 것입니다. 서로 비중이 다른 feature가 섞여서 weight에 대해 경합을 하면 오히려 좋지 않은 결과가 나오게 됩니다. 그래서, feature 갯수를 줄이는 것이 중요한데, 이 부분은 deep learning에서는 중요하지 않다고 볼 수 있습니다. deep learning은 sigmoid 대신 RELU 함수 계열을 사용하는 것도 있고 바로 뒤에 나오는 dropout을 통해 feature를 스스로 줄일 수 있는 방법도 있기 때문에 그렇습니다.

세 번째 : regularization.  weight이 너무 큰 값을 갖지 못하도록 제한하기 위한 방법입니다. 이 방법을 사용하면 weight이 커지지 않기 때문에 선이 구부러지는 형태를 피할 수 있다.

### Regularization

\begin{align} J(\theta) = \sum_{i=1} (h_\theta(x_i)-y_i)^2 + \lambda ||\theta||^2 \end{align}

\begin{align} h_\theta(x_i) = \theta^Tx_i + b \end{align}

\begin{align} \theta^* = \arg\min_\theta J(\theta) \end{align}
